# **Task 1: Conceptual Understanding**

**1. Difference between "Love" and "love" in NLP?**

In NLP, "Love" and "love" are considered different because computers treat uppercase and lowercase words separately. But in real meaning, both are the same. So we usually convert everything to lowercase to avoid confusion and keep the data clean.


**2. What happens if stopwords are not removed?**

If we don’t remove stopwords like is, the, and, the data becomes a bit noisy because these words don’t add much meaning. They can also slow down the model. But sometimes they are important depending on the task.


**3. Two real-world scenarios where removing stopwords can be harmful.**

Sentiment analysis
Example: “This is not good”
If we remove "not", the sentence becomes “good”, which gives the wrong meaning.
Search or questions
Example: “To be or not to be”
Removing stopwords will break the sentence and make it lose its actual meaning.


**4. Difference between stemming and lemmatization**

**Stemming** just cuts words to a shorter form, but sometimes the result is not a proper word.
Example: “running” → “run”, but “studies” → “studi”

**Lemmatization** is smarter. It converts words into their correct base form using meaning.
Example: “running” → “run”, “better” → “good”

So, stemming is faster but less accurate, while lemmatization is slower but gives better results.

#**Task 2:Build Advanced Preprocessing Function**

```python
import re

def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove email-like patterns
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Remove special characters (IMPORTANT FIX)
    text = re.sub(r"[^a-z\s]", " ", text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    #Remove words with length ≤ 2
    tokens = [word for word in tokens if len(word) > 2 or word in ("no", "not")]

    # Recreate cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

# **Task 3: Stress Testing**

```python
import re

def preprocess_text(text):
    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 4. Handle repeated characters (collapsing to single char per your example)
    text = re.sub(r'(.)\1+', r'\1', text)

    # 5. Remove punctuation and Emojis (keeping only letters and spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # 6. Remove short tokens (len <= 2) but keep "no" and "not"
    keep_words = {'no', 'not'}
    tokens = [t for t in text.split() if len(t) > 2 or t in keep_words]

    # 7. Final clean sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence

samples = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in samples:
    tokens, cleaned = preprocess_text(s)
    print(f"Original Text: {s}")
    print(f"Cleaned Tokens: {tokens}")
    print(f"Cleaned Sentence: {cleaned}")
    print("-" * 30)
```

# **Task 4: Token Analytics**

```python
def token_analysis(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    # Calculate average; handle potential division by zero for empty strings
    avg_len = sum(len(word) for word in tokens) / total if total > 0 else 0
    return total, unique, avg_len

print("\n========== TOKEN ANALYSIS ==========")

for s in samples:

    tokens, cleaned = preprocess_text(s)

    total, unique, avg_len = token_analysis(tokens)

    print(f"Original Text: {s}")
    print(f"Total tokens: {total}")
    print(f"Unique tokens: {unique}")
    print(f"Avg token length: {round(avg_len, 2)}")
    print("-" * 30)
```

**Analysis Questions**

**1. Which sentence had the most noise?**

**Call me at 9876543210" or "OK OK OK I got it"**.

The phone number example loses almost everything to the number-removal and short-token rules. The "OK" sentence has its primary content removed because "ok" is $\le 2$ characters and not in your "keep" list.

**2.Which sentence retained the most meaningful tokens?**

**"I absolutely looooved this product".**

After cleaning, it keeps high-value, descriptive words like "absolutely" and "loved." It also results in the highest Average Token Length (7.0), which is a strong indicator of "content-heavy" words versus short "filler" words.

# **Task 5: Frequency Analysis**

```python
from collections import Counter

# 1. Combine all tokens from all samples into one list
all_tokens = []
for s in samples:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

# 2. Perform Frequency Analysis
word_counts = Counter(all_tokens)

# 3. Identify Top 10 most frequent words
top_10 = word_counts.most_common(10)

# 4. Identify Top 5 least frequent words
# We sort the items by count (ascending)
least_5 = sorted(word_counts.items(), key=lambda x: x[1])[:5]

print("========== FREQUENCY ANALYSIS ==========")
print(f"Top 10 Most Frequent: {top_10}")
print(f"Top 5 Least Frequent: {least_5}")
```

#**Task 6: Build Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    all_clean_sentences = []

    for text in text_list:
        # Reusing the preprocess_text function logic
        tokens, clean_sentence = preprocess_text(text)

        # Flattening all tokens into one master list
        all_tokens.extend(tokens)

        # Storing each individual cleaned sentence
        all_clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": all_clean_sentences
    }

# Execution
results = full_pipeline(samples)

# Displaying Output
import json
print(json.dumps(results, indent=4))
```
